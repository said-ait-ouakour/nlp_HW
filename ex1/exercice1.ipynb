Implement the edit_one_letter function to get all the possible edits that are one edit away from a word.
The edits consist of the replace,
insert, delete, and optionally the switch operation

```python
def edit_one_letter(word,allow_switches = True):
        edit_list=[]
        edit_list=np.concatenate((insert_letter(word),replace_letter(word)))
        edit_list=np.concatenate((edit_list,delete_letter(word)))
        if allow_switches:
            edit_list=np.concatenate((edit_list,switch_letter(word)))
        edit_one_set = set(edit_list)
        return edit_one_set   
```

Implement the edit_two_letters function that returns a set of words that are two edits away    

```python
def edit_two_letters(word, allow_switches = True):
    """ edit with two letters

    Args:
        word (_type_): the word to be edited
        allow_switches (bool, optional): allow switches in the edit. Defaults to True.

    Returns:
        _type_: the set of known words with two letters edited.
    """
    edit_two_set = set()    
    edit_one = edit_one_letter(word,allow_switches=allow_switches)
    for w in edit_one:
        if w:
            edit_two = edit_one_letter(w,allow_switches=allow_switches)
            edit_two_set.update(edit_two)

    return edit_two_set
```

```python
  
def get_corrections(word, vocab, n:int=None):
    """ suggestions of corrections for a given word

    Args:
        word (str): a word to be corrected
        vocab (dict): vocab for the known words
        n (int) : the suggested number of corrections

    Returns:
        (list): list of corrections
    """
    suggestions = []
    suggestions = list((word in vocab and word) or edit_one_letter(word).intersection(vocab) or edit_two_letters(word).intersection(vocab))

    if n:
        return suggestions[:n]
    return suggestions
```

```python
def file_to_dictionary(file_path):

    dictionary = {}

    with open(file_path, 'r') as file:
        for line in file:
            if line:
                if(len(line.split())==2):
                    key, value = line.split()
                    dictionary[key] = int(value)
                columns=line.split()
                if len(columns)==3:
                    key=columns[0]+columns[1]
                    dictionary[key]=int(columns[2])
                
                if len(columns)==4:
                    key=columns[0]+columns[1]+columns[2]
                    dictionary[key]=int(columns[3] )   

    return dictionary    
```

```python

def min_edit_distance(source, target, ins_cost = 1, del_cost = 1, rep_cost = 2):
        """minimum edit distance function

        Args:
            source (str): the source word 
            target (str): the target word based on to edit 
            ins_cost (int, optional): the insertion cost. Defaults to 1.
            del_cost (int, optional): deletion cost. Defaults to 1.
            rep_cost (int, optional): replacing cost. Defaults to 2.

        Returns:
            int: minimum distance between two words
        """
        
        m = len(source) 
        n = len(target) 
        D = np.zeros((m+1, n+1), dtype=int) 
    
    
        for row in range(1,m+1): # Replace None with the proper range
            D[row,0] = D[row-1,0] + del_cost
        
        for col in range(1,n+1): # Replace None with the proper range
            D[0,col] = D[0,col-1] + ins_cost
        
        for row in range(1,m+1): 
        
            for col in range(1,n+1):
            
                r_cost = rep_cost
          
                if source[row-1] == target[col-1]:
                    r_cost = 0
                
           
                D[row,col] = min([D[row-1,col]+del_cost, D[row,col-1]+ins_cost, D[row-1,col-1]+r_cost])
          
        med = D[m,n]
    
        return  med
    
    
    
liste_corr_words=[]    
for i in sentence.split():
    tmp_corrections = get_corrections(i,  vocab)
    min_edit=min_edit_distance(i,tmp_corrections[0])
    correct_word=tmp_corrections[0]
    for target in tmp_corrections[1:]:
        minumum_edit= min_edit_distance(i,target)
        if(minumum_edit<=min_edit):
            min_edit=minumum_edit
            correct_word=target

            
    liste_corr_words.append(correct_word) 
print(liste_corr_words)
   
print("your corrected sentence is :  "+" ".join(liste_corr_words))  
```

```python
# UNQ_C1 (UNIQUE CELL IDENTIFIER, DO NOT EDIT)
### GRADED_FUNCTION: split_to_sentences ###
import re

def split_to_sentences(data):
    """
    Split data by linebreak "\n"
    
    Args:
        data: str
    
    Returns:
        A list of sentences
    """
    ### START CODE HERE (Replace instances of 'None' with your code) ###
    sentences = data.split("\n")
    ### END CODE HERE ###
    
    # Additional clearning (This part is already implemented)
    # - Remove leading and trailing spaces from each sentence
    # - Drop sentences if they are empty strings.
    sentences = [s.strip() for s in sentences]
    sentences = [s for s in sentences if len(s) > 0]
    
    return sentences
```

```python
# UNQ_C2 (UNIQUE CELL IDENTIFIER, DO NOT EDIT)
### GRADED_FUNCTION: tokenize_sentences ###
def tokenize_sentences(sentences):
    """
    Tokenize sentences into tokens (words)
    
    Args:
        sentences: List of strings
    
    Returns:
        List of lists of tokens
    """
    
    # Initialize the list of lists of tokenized sentences
    tokenized_sentences = []
    ### START CODE HERE (Replace instances of 'None' with your code) ###
    
    # Go through each sentence
    for sentence in sentences:
        
        # Convert to lowercase letters
        sentence = sentence.lower()
        
        # Convert into a list of words
        tokenized =  nltk.word_tokenize(sentence)
        tokenized = [x for x in tokenized if x != '']
         
        
        # append the list of words to the list of lists
        tokenized_sentences.append(tokenized)
    
    ### END CODE HERE ###
    
    return tokenized_sentences
```

```python
# UNQ_C4 (UNIQUE CELL IDENTIFIER, DO NOT EDIT)
### GRADED_FUNCTION: count_words ###
def count_words(tokenized_sentences):
    """
    Count the number of word appearence in the tokenized sentences
    
    Args:
        tokenized_sentences: List of lists of strings
    
    Returns:
        dict that maps word (str) to the frequency (int)
    """
        
    word_counts = {}
    ### START CODE HERE (Replace instances of 'None' with your code) ###
    
    # Loop through each sentence
    for sentence in tokenized_sentences: # complete this line
        
        # Go through each token in the sentence
        for token in sentence: # complete this line

            # If the token is not in the dictionary yet, set the count to 1
            if token not in word_counts: # complete this line
                word_counts[token] = 1
            
            # If the token is already in the dictionary, increment the count by 1
            else:
                word_counts[token] += 1

    ### END CODE HERE ###
    
    return word_counts
```

```python
# UNQ_C5 (UNIQUE CELL IDENTIFIER, DO NOT EDIT)
### GRADED_FUNCTION: get_words_with_nplus_frequency ###
def get_words_with_nplus_frequency(tokenized_sentences, count_threshold):
    """
    Find the words that appear N times or more
    
    Args:
        tokenized_sentences: List of lists of sentences
        count_threshold: minimum number of occurrences for a word to be in the closed vocabulary.
    
    Returns:
        List of words that appear N times or more
    """
    # Initialize an empty list to contain the words that
    # appear at least 'minimum_freq' times.
    closed_vocab = []
    
    # Get the word couts of the tokenized sentences
    # Use the function that you defined earlier to count the words
    word_counts = count_words(tokenized_sentences)
    
    ### START CODE HERE (Replace instances of 'None' with your code) ###

    # for each word and its count
    for word, cnt in word_counts.items(): # complete this line
        
        # check that the word's count
        # is at least as great as the minimum count
        if cnt>=count_threshold:
            
            # append the word to the list
            closed_vocab.append(word)
    ### END CODE HERE ###
    
    return closed_vocab
```

```python
def replace_oov_words_by_unk(tokenized_sentences, vocabulary, unknown_token="<unk>"):
    """
    Replace words not in the given vocabulary with '<unk>' token.
    
    Args:
        tokenized_sentences: List of lists of strings
        vocabulary: List of strings that we will use
        unknown_token: A string representing unknown (out-of-vocabulary) words
    
    Returns:
        List of lists of strings, with words not in the vocabulary replaced
    """
    
    # Place vocabulary into a set for faster search
    vocabulary = set(vocabulary)
    
    # Initialize a list that will hold the sentences
    # after less frequent words are replaced by the unknown token
    replaced_tokenized_sentences = []
    
    # Go through each sentence
    for sentence in tokenized_sentences:
        
        # Initialize the list that will contain
        # a single sentence with "unknown_token" replacements
        replaced_sentence = []
        ### START CODE HERE (Replace instances of 'None' with your code) ###

        # for each token in the sentence
        for token in sentence: # complete this line
            
            # Check if the token is in the closed vocabulary
            if token in vocabulary: # complete this line
                # If so, append the word to the replaced_sentence
                replaced_sentence.append(token)
            else:
                # otherwise, append the unknown token instead
                replaced_sentence.append(unknown_token)
        ### END CODE HERE ###
        
        # Append the list of tokens to the list of lists
        replaced_tokenized_sentences.append(replaced_sentence)
        
    return replaced_tokenized_sentences
```

```python
def extract_bigram_count(data):
    """ Extracts bigram counts from the preprocessed data

    Args:
        data (list): list of preprocessed data 

    Returns:
        dictionary: the bigram counts dictionary 
    """
    bigrams = defaultdict(int)  # Dictionary to store bigrams and their counts

    for sentence in data:
        for i in range(len(sentence) - 1):
            bigram = (sentence[i], sentence[i + 1])  # Create a pair of consecutive words
            bigrams[bigram] += 1  # Increment the count for the bigram

    return bigrams
```

```python
# UNQ_C8 (UNIQUE CELL IDENTIFIER, DO NOT EDIT)
### GRADED FUNCTION: count_n_grams ###
def count_n_grams(data, n, start_token='<s>', end_token = '<e>'):
    """
    Count all n-grams in the data
    
    Args:
        data: List of lists of words
        n: number of words in a sequence
    
    Returns:
        A dictionary that maps a tuple of n-words to its frequency
    """
    
    # Initialize dictionary of n-grams and their counts
    n_grams = {}
    
    for sentence in data: # complete this line
        
        # prepend start token n times, and  append <e> one time
        sentence = [start_token] * n+ sentence + [end_token]
        
        # convert list to tuple
        # So that the sequence of words can be used as
        # a key in the dictionary
        sentence = tuple(sentence)
        
        # Use 'i' to indicate the start of the n-gram
        # from index 0
        # to the last index where the end of the n-gram
        # is within the sentence.
        m = len(sentence) if n==1 else len(sentence)-1
        for i in range(m): # complete this line

            # Get the n-gram from i to i+n
            n_gram = sentence[i:i+n]

            # check if the n-gram is in the dictionary
            if n_gram in n_grams.keys(): # complete this line
            
                # Increment the count for this n-gram
                n_grams[n_gram] += 1
            else:
                # Initialize this n-gram count to 1
                n_grams[n_gram] = 1
    
    return n_grams
```

```python

def estimate_probability(word, previous_n_gram, 
                         n_gram_counts, n_plus1_gram_counts, vocabulary_size, k=1.0):
    """
    Estimate the probabilities of a next word using the n-gram counts with k-smoothing
    
    Args:
        word: next word
        previous_n_gram: A sequence of words of length n
        n_gram_counts: Dictionary of counts of n-grams
        n_plus1_gram_counts: Dictionary of counts of (n+1)-grams
        vocabulary_size: number of words in the vocabulary
        k: positive constant, smoothing parameter
    
    Returns:
        A probability
    """
    # convert n_grams to tuple of words
    previous_n_gram = tuple(previous_n_gram)
   
    if previous_n_gram in n_gram_counts:
        previous_n_gram_count = n_gram_counts[previous_n_gram]
    else:
        previous_n_gram_count=0
        
    # using k-smoothing
    denominator = previous_n_gram_count+k*vocabulary_size

    # Define n plus 1 gram as the previous n-gram plus the current word as a tuple
    n_plus1_gram = previous_n_gram + (word,)
  
    # Set the count to the count in the dictionary,
    # otherwise 0 if not in the dictionary
    # use the dictionary that has counts for the n-gram plus current word
    if n_plus1_gram in  n_plus1_gram_counts:
        n_plus1_gram_count =  n_plus1_gram_counts[n_plus1_gram]
    else:
        n_plus1_gram_count=0
        
    # Define the numerator use the count of the n-gram plus current word,
    # and apply smoothing
    numerator = n_plus1_gram_count+k

    # Calculate the probability as the numerator divided by denominator
    probability = numerator/denominator
      
    return probability
```

```python
def file_to_dictionary(file_path):
    """convert file to dictionary

    Args:
        file_path (str): file path

    Returns:
        dict: dictionary of words 
    """
    dictionary = {}
    with open(file_path, 'r') as file:
        for line in file:
            if line:
                if(len(line.split())==2):
                    key, value = line.split()
                    dictionary[key] = int(value)
                columns=line.split()
                if len(columns)==3:
                    key=columns[0]+columns[1]
                    dictionary[key]=int(columns[2])
                
                if len(columns)==4:
                    key=columns[0]+columns[1]+columns[2]
                    dictionary[key]=int(columns[3] )   

    return dictionary

def get_probs(word_count_dict):
    """ get probabilities for word counter

    Args:
        word_count_dict (dict): word counter dictionary

    Returns:
        dict: probabilities 
    """
    probabilities={}
    size=sum(word_count_dict.values())
    for i in word_count_dict.keys():
        probabilities[i]=word_count_dict[i]/size
    return probabilities




def known(words,data): 
    "The subset of `words` that appear in the dictionary of WORDS."
    return [w for w in words if w in data]


def count_freq_letters(dictionary_big_data):
    """ count the frequencies of word in a given corpus

    Args:
        dictionary_big_data (dict): dictionary of WORDS

    Returns:
        dict: counter of words 
    """
    letters="abcdefghijklmnopqrstuvwxyz"
    dict_letters_big_data={}
    for letter in letters:
        count=0
        for word in dictionary_big_data.keys():
            for i in range(len(word)):
                if(letter==word[i]):
                    count+=dictionary_big_data[word]
        dict_letters_big_data[letter]=count  
        
    return dict_letters_big_data    
```

```python
#insert_list is a filtred list using big.txt
'''this function allows us to know which letter is inserted in the wrong word to calculate probability that a new letter written next to the old letter
for example :   the wrong word is acress and the suggested word is cress so we calculate prob(a|#) 
                 the wrong word is acress and the suggested word is acres so we calculate prob(es|e) or prob(ss|s)
                 return a list of inserted letters like ss|s es|e
'''
def estimate_words_by_insert(wrong_word,insert_list):
    word_list=[]
    leng_w1=len(wrong_word)
    for word in insert_list:
        leng_w2=len(word) 
        k=0
        if(word[0]!=wrong_word[0] or word[0]==word[1]==wrong_word[0]):
            mot=word[0]+"|#"
            word_list.append(mot)
            continue
                   
        for i in range(leng_w1):
            if(word[i]!=wrong_word[i]):
                k+=1
                mot=word[i-1]+"|"+word[i-1]+word[i]
                word_list.append(mot)
                break
                
        if(k==0):
            mot=word[leng_w2-2]+"|"+word[leng_w2-2]+word[leng_w2-1]
            word_list.append(mot)
            
    return word_list 
            
'''this function allows us to know which letter is deleted in the wrong word to calculate probability that a new letter writed next to the old letter
for example :   the wrong word is acress and the suggested word is actress so we calculate prob(c|ct) 
                 return a list of deleted letter like c|ct
'''       
def estimate_words_by_delete(wrong_word,delete_list):
    #filter
    word_list=[]
    leng_w1=len(wrong_word)
    for word in delete_list:
        leng_w2=len(word) 
        k=0
        if(word[0]!=wrong_word[0]):
            mot=wrong_word[0]+wrong_word[1]+"|"+word[0]
            word_list.append(mot)
            continue
                   
        for i in range(leng_w2):
            if(word[i]!=wrong_word[i]):
                k+=1
                mot=wrong_word[i]+wrong_word[i+1]+"|"+word[i]
                word_list.append(mot)
                
        if(k==0):
            mot=wrong_word[leng_w2-1]+wrong_word[leng_w2]+"|"+word[leng_w2-1]
            word_list.append(mot)
            
    return word_list 

'''this function allows us to know which letter is replaced in the wrong word to calculate probability that a new letter replaced by the other letter
for example :   the wrong word is acress and the suggested word is access so we calculate prob(r|c) 
              return replace list of letters like r|c   
'''
def estimate_words_by_replace(wrong_word,replace_list):
    replaced_list=[]
    for word in replace_list:
        for i in range(len(wrong_word)):
            if(wrong_word[i]!=word[i]):
                mot=wrong_word[i]+"|"+word[i]
                replaced_list.append(mot)
                
    return replaced_list  

'''this function allows us to know which letters are switched in the wrong word to calculate probability that a new letters are switched by the others letters
for example :   the wrong word is acress and the suggested word is caress so we calculate prob(ca|ac) 

return list of switched letters like ca|ac
                 
'''
def estimate_words_by_switch(wrong_word,switch_list):
    switched_list=[]
    for word in switch_list:
        if(word[0]!=wrong_word[0]):
            mot=word[1]+word[0]+"|"+word[0]+word[1]
            switched_list.append(mot)
            continue 
            
            for i in range(len(word)-1):
                if(word[i]!=wrong_word[i]):
                    mot=word[i]+word[i+1]+"|"+word[i+1]+word[i]
                    switched_list.append(mot)
            
    return switched_list


#to make a dict
def make_dict(liste_error,liste_corr_word):
    dicts={}
    if(len(liste_error)==1):
        dicts[liste_error[0]]=liste_corr_word[0]
    else:
        for i in range(len(liste_error)):
            dicts[liste_error[i]]=liste_corr_word[i]
            
    return dicts        
```

```python
import numpy as np

def make_final_dict_corr(word):
    insert_list=known(insert_letter(word),WORDS)
    corr_error_insert=np.unique(estimate_words_by_insert(word,insert_list))
    
    delete_list=known(delete_letter(word),WORDS)

    corr_error_delete=np.unique(estimate_words_by_delete(word,delete_list))

    replace_list=known(replace_letter(word),WORDS)

    corr_error_replace=np.unique(estimate_words_by_replace(word,replace_list))

    switch_list=known(switch_letter(word),WORDS)

    corr_error_switch=np.unique(estimate_words_by_switch(word,switch_list))

    dict1=make_dict(corr_error_insert,insert_list)
    dict2=make_dict(corr_error_delete,delete_list)
    dict3=make_dict(corr_error_replace,replace_list)
    dict4=make_dict(corr_error_switch,switch_list)
    corr_dict={}
    corr_dict.update(dict1)
    corr_dict.update(dict2)
    corr_dict.update(dict3)
    corr_dict.update(dict4)
    return corr_dict
      


```
